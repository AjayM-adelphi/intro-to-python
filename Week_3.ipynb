```python
import math

def sphere_volume(radius, *, digits=5):
    """Calculate the volume of a sphere.

    Args:
        radius (float): radius of the sphere
        digits (int, optional): number of digits for rounding

    Returns:
        float: volume of the sphere
    """
    volume = (4/3) * math.pi * radius**3
    return float(round(volume, digits))
```

```python
radius = 100.0

sphere_volume(radius, digits=1)
sphere_volume(radius, digits=5)
sphere_volume(radius, digits=10)
sphere_volume(radius, digits=15)
sphere_volume(radius, digits=20)
```

```python
radius = 42.0

for d in range(1, 21):
    print(sphere_volume(radius, digits=d))
```
