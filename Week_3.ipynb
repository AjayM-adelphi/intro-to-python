```python
import math

def sphere_volume(radius, *, digits=5):
    """Calculate the volume of a sphere.

    Args:
        radius (float): radius of the sphere
        digits (int, optional): number of digits for rounding

    Returns:
        float: volume of the sphere
    """
    volume = (4/3) * math.pi * radius**3
    return float(round(volume, digits))
```

```python
radius = 100.0

sphere_volume(radius, digits=1)
sphere_volume(radius, digits=5)
sphere_volume(radius, digits=10)
sphere_volume(radius, digits=15)
sphere_volume(radius, digits=20)
```

```python
def photon_energy(wavelength_nm):
    """
    Compute the energy of a photon given its wavelength in nanometers.

    Parameters
    ----------
    wavelength_nm : float
        Wavelength in nanometers.

    Returns
    -------
    float
        Photon energy in electron-volts (eV).
    """
    hc_ev_nm = 1240  # eV·nm
    return hc_ev_nm / wavelength_nm
```

```python
def photon_energy(wavelength_nm, conversion_factor=1):
    """
    Compute photon energy given its wavelength.

    Parameters
    ----------
    wavelength_nm : float
        Wavelength in nanometers.
    conversion_factor : float
        Multiply the result to convert units

    Returns
    -------
    float
        Photon energy in desired units.
    """
    hc_ev_nm = 1240  # eV·nm
    energy_ev = hc_ev_nm / wavelength_nm
    return energy_ev * conversion_factor

# 1.60218e-19 to return energy in Joules
energy_600nm_J = photon_energy(600, 1.60218e-19)
print(energy_600nm_J)
```

```python
from scipy.constants import h, c, e

def photon_energy(wavelength_nm, joules=False):
    """
    Compute photon energy using the built in scipy constants

    Parameters
    ----------
    wavelength_nm : float
        Wavelength in nanometers
    joules : bool
        If True, return energy in Joules.
        If False (default), return energy in eV
    Returns
    -------
    float
        Photon energy in Joules or eV.
    """
    wavelength_m = wavelength_nm * 1e-9
    energy_joules = (h * c) / wavelength_m

    if joules:
        return energy_joules
    else:
        return energy_joules / e


#example

photon_energy(1000)            # eV
photon_energy(1000, True)      # Joules
```

```python
# 03_conditials

def check_uncertainty_principle(delta_x, delta_p):
    """Check if a measurement satisfies the Heisenberg uncertainty principle

    Args:
        delta_x (float): uncertainty in position
        delta_p (float): uncertainty in momentum kg/m*s

    Returns:
        bool: True if measurement is physically possible, False otherwise
    """
    hbar = 1.055e-34  # J·s
    return delta_x * delta_p >= hbar / 2
```

```python
# Case a
check_uncertainty_principle(1e-10, 1e-24)

# Case b
check_uncertainty_principle(1e-10, 1e-26)
# Case c
check_uncertainty_principle(1e-15, 1e-19)

# For some reason to check you have to do the code one at a time, but it  looks like the order is True, False, True
```

```python
# The explanation why case b is false is because in quantum mechanics you cannot know the momentum and position too precisely and thus the measurements in this case violate the uncertainty principle.
```

```python
# For this question we need to re write the formula in a way to get eV

def can_eject_electron(wavelength, work_function):
    """Determine if a photon can eject an electron from a object"""
    photon_energy = 1240 / wavelength
    return photon_energy >= work_function
```

```python
def electron_kinetic_energy(wavelength, work_function):
    """Calculate kinetic energy of ejected electron"""
    photon_energy = 1240 / wavelength

    if photon_energy < work_function:
        return 0.0

    return photon_energy - work_function
```

```python
wavelength = 400

# Sodium
work_function_na = 2.75
print("Sodium - Ejection occurs:", can_eject_electron(wavelength, work_function_na))
print("Sodium - KE:", electron_kinetic_energy(wavelength, work_function_na), "eV")

# Aluminum
work_function_al = 4.28
print("Aluminum - Ejection occurs:", can_eject_electron(wavelength, work_function_al))
print("Aluminum - KE:", electron_kinetic_energy(wavelength, work_function_al), "eV")

# Copper
work_function_cu = 4.65
print("Copper - Ejection occurs:", can_eject_electron(wavelength, work_function_cu))
print("Copper - KE:", electron_kinetic_energy(wavelength, work_function_cu), "eV")

# Gold
work_function_au = 5.1
print("Gold - Ejection occurs:", can_eject_electron(wavelength, work_function_au))
print("Gold - KE:", electron_kinetic_energy(wavelength, work_function_au), "eV")
```

```python
def analyze_photon(wavelength, work_function):
    """Provide comprehensive analysis of photon-material interaction."""

    # Out of range
    if wavelength <= 10 or wavelength >= 1_000_000:
        return f"Wavelength out of range ({wavelength} nm). Cannot classify wavelength"

    # Spectral classification
    if wavelength < 380:
        region = "UV"
    elif wavelength <= 700:
        region = "Visible"
    else:
        region = "IR"

    photon_energy = 1240 / wavelength

    if photon_energy >= work_function:
        ke = photon_energy - work_function
        return f"{region} photon ({wavelength} nm)."
    else:
        return f"{region} photon ({wavelength} nm). No electron ejection (insufficient energy)"

```

```python
def minimum_momentum_uncertainty(delta_x):
    hbar = 1.055e-34
    return hbar / (2 * delta_x)
```
